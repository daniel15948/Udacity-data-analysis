### <a name="orginal">Explore data and plots </a>
From data set global average temperature data start at 1750 and end of 2015,most of average temperature not over 10 degrees celsius.From my live city of San Francisco temperature data set start at 1849 and end of 2013.most of average temperature between 14 - 16 degree celsius.

````python
# Global temperature average data set
global_data = pd.read_csv('global_results.csv')

# San Francisco temperture average data set
sf_data = pd.read_csv('sanfrancisco_results.csv')
````

### Plots

````python
plt.figure(figsize=(20,8))
plt.plot(global_data.year, global_data.avg_temp,label='Global_avg_temp')
plt.axhline(global_data.avg_temp.mean(), color='k', linestyle='dashed', linewidth=2,label='Global_avg_temp_mean')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Temperature(℃)')
plt.title('Global Weather Trends')
plt.xticks(np.arange(1750, 2016, step=5), rotation=40)
plt.show()
````

From global average temperature trends fold line,as we can see the global temperature on the up trend.

````python
plt.figure(figsize=(20,8))
plt.plot(sf_data.year, sf_data.avg_temp,color='orange',label='sf_avg_temp')
plt.axhline(sf_data.avg_temp.mean(), color='k', linestyle='dashed', linewidth=2,label='sf_avg_temp_mean')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Temperature(℃)')
plt.title('San Francisco Weather Trends')
plt.xticks(np.arange(1849, 2016, step=5), rotation=40)
plt.show()
````

To observation San Francisco average temperature trends,the temperature looks not too much rising trends.it's worth to analyze the trough temperature average and peak temperature average or maybe to find some interesting values in temperature moving average data set. 

````python
# San Francisco average temperature compare with global average temperaure

plt.figure(figsize=(20,8))
plt.plot(global_data.year, global_data.avg_temp, label='Global')
plt.plot(sf_data.year, sf_data.avg_temp, label='San Francisco')
plt.legend()
plt.title('San Francisco vs Global Weather Trends')
plt.xlabel('Year')
plt.ylabel('Temperature(℃)')
plt.xticks(np.arange(1750, 2016, step=5), rotation=40)
plt.show()
````

To observation San francisco and global weather trends fold lines,it's hard to anaylze weather trends.Because of the fold line fluctuates by years,with moving temperature could be solve the problem.

### <a name="moving">Moving average and plots </a>

Setting temperature moving average of 5 years,7years and 10 years then saving into "global_data" and "sf_data"

````python
n = 5
global_data['moving_avg5'] = global_data['avg_temp'].rolling(window=n).mean()
sf_data['moving_avg5'] = sf_data['avg_temp'].rolling(window=n).mean()

n = 7
global_data['moving_avg7'] = global_data['avg_temp'].rolling(window=n).mean()
sf_data['moving_avg7'] = sf_data['avg_temp'].rolling(window=n).mean()

n = 10
global_data['moving_avg10'] = global_data['avg_temp'].rolling(window=n).mean()
sf_data['moving_avg10'] = sf_data['avg_temp'].rolling(window=n).mean()
````

````python
global_data.head(10)
````

````python
sf_data.head(10)
````

### Plots

````python
plt.figure(figsize=(20,8))
plt.plot(global_data.year, global_data.avg_temp, label='year_avg')
plt.plot(global_data.year, global_data.moving_avg5, label='moving_avg5')
plt.plot(global_data.year, global_data.moving_avg7, label='moving_avg7')
plt.plot(global_data.year, global_data.moving_avg10, label='moving_avg10')
plt.legend()

plt.title('Global Weather Trends')
plt.xlabel('Year')
plt.xticks(np.arange(1750, 2016, step=5), rotation=40)
plt.ylabel('Temperature(℃)')
plt.show()
````

From 5 years,7years and 10 years global moving average.those fold lines are more clearly to understand the average temperature trends.

````python
plt.figure(figsize=(20,8))
plt.plot(sf_data.year, sf_data.avg_temp, label='year_avg')
plt.plot(sf_data.year, sf_data.moving_avg5, label='moving_avg5')
plt.plot(sf_data.year, sf_data.moving_avg7, label='moving_avg7')
plt.plot(sf_data.year, sf_data.moving_avg10, label='moving_avg10')
plt.legend()
plt.title('San Francisco Weather Trends')
plt.xlabel('Year')
plt.xticks(np.arange(1849, 2016, step=5), rotation=40)
plt.ylabel('Temperature(℃)')
plt.show()
````

Compared with San Francisco year temperature average and moving average fold lines,moving average fold lines are obviously much understanding than year average.Next,Let's pick up 10 years moving average to find the weather tends.

### <a name="moveplots">Moving average data compare plots </a>
The reason for choiced 10 years fold line for each data because San Francisco 10 years moving average temperature data start from 1849,so I added a new variate global_data2 let's the data start from 1849 as well. 

````python
global_data2 = global_data[global_data.year >= 1849]
````

````python
# global 10 years moving average temperature
plt.figure(figsize=(20,8))
plt.plot(global_data2.year, global_data2.moving_avg10, label='Global', color='steelblue', linewidth=3)
plt.title('Global 10 Years Moving Average Temperature')
plt.xlabel('Year')
plt.ylabel('Temperature(℃)')
plt.xticks(np.arange(1850, 2016, step=5), rotation=40)
plt.legend()
plt.show()

# San Francisco 10 years moving average temperature
plt.figure(figsize=(20,8))
plt.plot(sf_data.year, sf_data.moving_avg10, label='San Francisco', color='orangered', linewidth=3)
plt.title('San Francisco 10 Years Moving Average Temperature')
plt.xlabel('Year')
plt.ylabel('Temperature(℃)')
plt.xticks(np.arange(1850, 2016, step=5), rotation=40)
plt.legend()
plt.show()
````

````python
global_data2['moving_avg10'].describe()
````

````python
sf_data['moving_avg10'].describe()
````

Combine with global 10 years moving average and San Francisco 10 years moving average fold lines together.

````python
# san francisco vs global moving average temperture 
plt.figure(figsize=(20,10))
plt.plot(global_data2.year, global_data2.moving_avg10, label='Global', color='steelblue', linewidth=3)
plt.plot(sf_data.year, sf_data.moving_avg10, label='San Francisco', color='orangered', linewidth=3)
plt.legend()
plt.title('San Francisco vs Global 10 Years Moving Average Temperature')
plt.xlabel('Year')
plt.ylabel('Temperature(℃)')
plt.xticks(np.arange(1850, 2016, step=5), rotation=40)
plt.show()
````
